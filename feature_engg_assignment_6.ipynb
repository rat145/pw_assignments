### Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.


```python
import pandas as pd
import numpy as np

study_time = np.random.randint(1, 10, size=100)
exam_scores = np.random.randint(50, 100, size=100)
df = pd.DataFrame({
    'study_time': study_time,
    'exam_scores': exam_scores
})

df.head()
```

```python
df['study_time'].corr(df['exam_scores'], method = "pearson")
```

```python
'''
1. The correlation of study time to exam scores is positively correlated.
2. Which means as the study time increases, the exam scores also increases.
'''
```

### Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables. Suppose you have collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two variables and interpret the result.

```python
sleep_time = np.random.randint(5, 10, size=100)
job_satisfaction = np.random.randint(1, 11, size=100)
df = pd.DataFrame({'sleep_time': sleep_time, 'job_satisfaction': job_satisfaction})
df
```

```python
df['sleep_time'].corr(df['job_satisfaction'], method = "spearman")
```

```python
'''
There is a weak negative correlation between the sleep time and job satisfaction.
According to the output it can be said that people who get more sleep are less satisfied with job.
However this is a weak negative correlation.
'''
```

### Q3. Suppose you are conducting a study to examine the relationship between the number of hours of exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation between these two variables and compare the results.

```python
# Generate random values for each variable
exercise_time = np.random.randint(0, 10, size=50)
bmi = np.random.randint(18, 35, size=50)

# Create a Pandas DataFrame from the data
df = pd.DataFrame({'exercise_time': exercise_time, 'bmi': bmi})

df.head()
```

```python
df['exercise_time'].corr(df['bmi'], method = "pearson")
```

```python
df['exercise_time'].corr(df['bmi'], method = "spearman")
```

```python
'''
1. The pearson correlation is slighlty higher than spearman correlation.
2. Both the techniques are used to see relationships between variables.
3. However they are calculated differently.
4. Pearson correlation gives us linear relationship.
5. Spearman correlation gives us monotonic relationship.
6. Final interpretation would be that people who exercise more tend to have
lower BMI.
'''
```

### Q4. A researcher is interested in examining the relationship between the number of hours individuals spend watching television per day and their level of physical activity.
