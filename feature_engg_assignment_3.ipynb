### Q1. What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its application.

```python
'''
1. Min Max scaler is a normalization technique used in datasets to scale data.
2. It's value lies from 0 to 1.
3. The more the value is towards 0, the more it is close to minimum value.
4. The more the value is towards 1, the more it is close to maximum value.
'''
```

```python
# importing min max scaler from sklearn.preprocessing library
from sklearn.preprocessing import MinMaxScaler
```

```python
# creating an instance of minmaxscaler
min_max = MinMaxScaler()
```

```python
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

```python
# Lets apply normalization on total_bill feature.
min_max.fit_transform(df[['total_bill']])
```

```python
bill_normalized_array = min_max.fit_transform(df[['total_bill']])
```

```python
import pandas as pd
# Lets add it in the dataframe
df_normalized = pd.DataFrame(bill_normalized_array, columns = ['total_bill_scaled'])
df_normalized.head()
```

```python
pd.concat([df,df_normalized], axis = 1)
```

### Q3. What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an example to illustrate its application.

```python
'''
**Principal Component Analysis (PCA)** is a statistical technique used for **dimensionality reduction**Â¹. It aims to transform a dataset with a large number of variables into a new set of variables called **principal components**Â¹. These principal components are linear combinations of the original variables and are ordered in decreasing order of importanceÂ¹.

PCA is used to reduce the dimensionality of a dataset while preserving the most important patterns or relationships between the variablesÂ¹. It is widely used in exploratory data analysis and machine learning for predictive modelsÂ¹. By reducing the number of dimensions, PCA helps to address the **curse of dimensionality**, which refers to issues such as overfitting, increased computation time, and reduced accuracy that arise when working with high-dimensional dataÂ¹.

Here's an example to illustrate how PCA works. Let's say we have a dataset with three features: `x`, `y`, and `z`. We want to reduce the dimensionality of this dataset using PCA. The original dataset looks like this:

| **x** | **y** | **z** |
|-------|-------|-------|
| 2     | 4     | 6     |
| 3     | 6     | 9     |
| 4     | 8     | 12    |

To apply PCA, we first calculate the covariance matrix of the original dataset. Then, we find the eigenvectors and eigenvalues of this covariance matrix. The eigenvectors represent the principal components, and the eigenvalues represent the amount of variance explained by each principal component.

In this example, let's assume that the eigenvectors and eigenvalues are as follows:

- Eigenvector 1: [0.2673, 0.5345, 0.8018]
- Eigenvector 2: [0.5345, -0.8018, 0.2673]
- Eigenvector 3: [0.8018, 0.2673, -0.5345]
- Eigenvalue 1: 30
- Eigenvalue 2: 10
- Eigenvalue 3: 5

We can see that the first eigenvector (PC1) captures the maximum amount of variance in the data, followed by PC2 and PC3. Therefore, we can choose to retain only PC1 if we want to reduce the dimensionality to one.

The transformed dataset using PCA would look like this:

| **PC1** |
|---------|
| -2.6832 |
| -5.3664 |
| -8.0496 |

As you can see, we have successfully reduced the dimensionality from three features (`x`, `y`, `z`) to one principal component (`PC1`).
'''
```

### Q6. You are working on a project to build a model to predict stock prices. The dataset contains many features, such as company financial data and market trends. Explain how you would use PCA to reduce the dimensionality of the dataset.

```python
import pandas as pd
import numpy as np

# Create a dictionary with the data
data = {'Company': ['A', 'B', 'C', 'D'],
        'Revenue': np.random.randint(100000, 1000000, size=4),
        'Profit': np.random.randint(10000, 100000, size=4),
        'Market Share': np.random.uniform(0.1, 0.5, size=4),
        'Growth Rate': np.random.uniform(0.01, 0.1, size=4)}

# Create a pandas DataFrame from the dictionary
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

features_to_scale = ['Revenue', 'Profit', 'Market Share', 'Growth Rate']
scaler = StandardScaler()
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])
```

```python
pca = PCA(n_components=2)
pca.fit(df[features_to_scale])
```

```python
transformed_data = pca.transform(df[features_to_scale])
```
