### Q1. Explain GET and POST methods.

- GET and POST methods are methods in HTML forms.
- When the data is displayed on the address bar, it is GET method.
- When the data is hidden from the address bar, it is POST method.

### Q2. Why is request used in Flask?

- request is used when we want to catch data from HTML document.
- Usually we catch data from the forms in HTML.
- Through HTML forms we can take user input and catch those data using request.
- The data if taken through forms, by default it is in string data type.
- We can perform logical operations, or any operation we want to perform using request.

### Q3. Why is redirect() used in Flask?

- redirect function is used to send a user to different url.
- We can use redirect to send a user to its profile page after he is finished logging in.
- It is a very important feature since it can be used to redirect users to protected web pages.
- For example after logging in to PW Skills website when OTP is validated I am redirected to my course dashboard which no other user can visit.

### Q4. What are templates in Flask? Why is the render_template() function used?

- templates is a folder which contains our HTML files.
- The folder should be named templates and should be saved in the current working directory.
- We can use render_template() function to run the HTML file and display it on the web app.
- This function renders the HTML document and opens it.
- For example: render_template("index.html")
