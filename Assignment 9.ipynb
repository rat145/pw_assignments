### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

- The LookupError class is used in python to address the errors where there was an error in looking up the value in a collection.
- Its subclasses include KeyError and IndexError.

1. KeyError

When in a dictionary, we are trying to find a value using key, but the key does not exist in the dictionary then it will throw KeyError.

2. IndexError

When in a list, we are trying to find an element with a specific index, but the index we mentioned does not have any item then it will throw IndexError.

```python
# KeyError

try:
    logging.info("inside try block defining a dictionary variable")
    d = {"one":1, "two":2, "three":3}
    logging.info("trying to find a key value pair using key")
    d["four"]
except KeyError as e:
    logging.info("inside except block, key error")
    print("This key doesn't exist",e)
```

```python
# IndexError

try:
    logging.info("inside try block defining a list variable")
    l = [1,2,3,4,5,564,2.4,(2,"pw"),"rat",[True,False,"ppp"]]
    logging.info("trying to find an element using index")
    l[10]
except IndexError as e:
    logging.info("inside except block, index error")
    print(e)
```
