### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask:
- Flask framework is a Web API package for python.
- It used to send and receive HTTP requests and responses.

Advantages of Flask:
- It is less complex which makes it easy to understand.
- It does not impose many restrictions.
- We can use any component and integrate third-party libraries.
- It has in-built development server so we can test our API locally without having to set up anything differently.
- We can use routing function to build URLs.
