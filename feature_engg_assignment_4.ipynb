### Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario.

````python
'''
1. Nominal Encoding or OHE (One Hot Encoding) is a technique used to transfer
categorical variables that have no intrinsic ordering into numerical values that
can be used in machine learning models.
2. One common method for nominal encoding is one-hot encoding, which creates a
binary vector for each category in the variable.
'''
````

````python
from sklearn.preprocessing import OneHotEncoder
import seaborn as sns
import pandas as pd
````

````python
encoder = OneHotEncoder()
````

````python
df = sns.load_dataset("tips")
df.head()
````

````python
df['time'].unique()
````

````python
encoded_data = encoder.fit_transform(df[['time']]).toarray()
````

````python
encoded_df = pd.DataFrame(encoded_data, columns = ['Dinner', 'Lunch'])
encoded_df.head()
````

````python
pd.concat([df, encoded_df], axis = 1)
````

### Q3. In what situations is nominal encoding preferred over one-hot encoding? Provide a practical example.

````python
'''
Nominal encoding, also known as label encoding, is preferred over one-hot encoding
in situations where:

1. Ordinal Relationship Exists: Nominal encoding is suitable when there is a
natural ordinal relationship among the categories, meaning that the categories
have an inherent order or hierarchy. In such cases, assigning numerical labels can
capture this order more effectively than one-hot encoding.

2. Reduced Dimensionality: Nominal encoding reduces the dimensionality of the data
compared to one-hot encoding. This can be beneficial when dealing with large
categorical feature spaces, as it saves memory and computational resources.

3. Improved Interpretability: For some machine learning algorithms, nominal
encoding may lead to more interpretable models. Some algorithms, like decision
trees and random forests, can work well with numerical labels because they can
naturally split and make decisions based on the numeric values.

Here's a practical example to illustrate when nominal encoding is preferred over
one-hot encoding:

Example: Education Levels

Suppose you have a dataset containing information about individuals, including
their education levels. The education levels can be categorized as "High School,"
"Bachelor's Degree," "Master's Degree," and "Ph.D." These education levels have an
inherent order, with "Ph.D." being the highest and "High School" being the lowest.
In this case:

- Nominal Encoding: You assign numerical labels to the education levels, such
as 1 for "High School," 2 for "Bachelor's Degree," 3 for "Master's Degree," and 4 for
"Ph.D." This preserves the ordinal relationship among education levels while reducing
dimensionality.

- One-Hot Encoding: If you one-hot encode this categorical feature, you would create
four binary columns, one for each education level. However, this doesn't capture the
ordinal relationship among the levels, and it increases dimensionality.

Using nominal encoding in this scenario allows you to represent the education levels
in a way that reflects their natural order. It's more efficient in terms of
dimensionality and can lead to a more interpretable model if you're using algorithms
that work well with ordinal data.

Keep in mind that the choice between nominal encoding and one-hot encoding depends on
the specific characteristics of your data and the requirements of your machine learning
algorithm. In cases where there is no ordinal relationship among categories, and each
category is equally important, one-hot encoding is often the preferred choice to avoid
introducing unintended ordinal relationships.
'''
````

### Q5. In a machine learning project, you have a dataset with 1000 rows and 5 columns. Two of the columns are categorical, and the remaining three columns are numerical. If you were to use nominal encoding to transform the categorical data, how many new columns would be created? Show your calculations.

````python
import numpy as np
df = pd.DataFrame({
    "Gender": np.random.choice(['Male','Female'], size = 1000),
    "Hobby": np.random.choice(["Singing","Dancing","Football","Video Games"], size = 1000),
    "Time_studied": np.random.uniform(1,9, size = 1000),
    "Time_played": np.random.uniform(1,9, size = 1000),
    "Time_slept": np.random.uniform(4,12, size = 1000)
})
df.head()
````

````python
encoder = OneHotEncoder()
````

````python
encoded_data = encoder.fit_transform(df[['Gender','Hobby']]).toarray()
````

````python
encoded_data
````

````python
encoded_df = pd.DataFrame(encoded_data, columns = ['Female', 'Male', 'Dancing', 'Football', 'Singing', 'Video Games'])
encoded_df.head()
````

````python
pd.concat([df, encoded_df], axis = 1)
````

````python
'''
According to our categories it will be determined how many number of columns we are
going to have.
Here I had total 6 unique categories from both categorical columns and hence now I
have 6 more columns in addition to already existing 5.
'''
````
